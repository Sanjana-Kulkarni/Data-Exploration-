#1. Missing Values:
Question: Are there any missing values in the dataset, and if so, how should they be handled for each indicator?


  Answer: Yes, there are missing values in dataset(Age(7),Rating(1)).
   'Age' and 'Rating' can be handled using mean method
   

```python
data_frame.isnull().sum()
```

```python
data_frame['Age']
```

```python
avg_age = data_frame['Age'].mean()
avg_age
```

```python
# fill the missing values with mean
data_frame['Age'] = data_frame.Age.fillna(avg_age)
data_frame['Age'] = data_frame.Age.round(decimals=1)
```

```python
data_frame.head()
```

```python
data_frame['Rating']
```

```python
data_frame.loc[data_frame['Rating']<=0,'Rating']=np.nan

df_rating = data_frame['Rating'].mean()
df_rating
```

```python
data_frame['Rating']=data_frame['Rating'].fillna(df_rating)
```

```python
data_frame.Rating.round(decimals=0)
```

#4. Salary Formatting:

#5. Location Standardization:
Question: Check the consistency of location entries. Do they need standardization, and how can this be achieved?

Answer: Yes, they need to standardise


```python
data_frame.Location = data_frame.Location.str.replace(',In', ' ')
data_frame.Location = data_frame.Location.str.replace(' In', ' ')
data_frame.Location = data_frame.Location.str.replace('Australia Aus', 'Australia ')
data_frame.Location = data_frame.Location.str.replace(',Ny', ' ')
```

```python
data_frame
```

#6. Established Column:
Question: Explore the Established column. Are there any inconsistencies or anomalies that need to be addressed?

Answer: Yes, there are some inconsistencies that need to be addressed




```python
data_frame.loc[data_frame['Established']<=0,'Established']=np.nan

Year = data_frame['Established'].median()
Year
```

```python
data_frame['Established']=data_frame['Established'].fillna(Year).astype(int)
```

#9. Age Distribution:
Question: Check the distribution of values in the Age column. Are there any unusual entries, and how might they impact analysis?


```python
age_distribution = data_frame['Age'].value_counts().sort_index()
```

```python
age_distribution.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
